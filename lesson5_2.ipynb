```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
#fig = figure()
#ax = Axes3D(fig)
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)

Z = np.sqrt(5.*(X**2 + Y**2)/1. + 2)

xcolors = X - min(X.flat)
xcolors = xcolors/max(xcolors.flat)
surf = ax.plot_surface(X, Y, Z, facecolors=cm.hot(xcolors))

plt.show()
```
