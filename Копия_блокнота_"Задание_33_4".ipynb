```python
#решить систему уравнений:
#  exp(x) + x∙(1 – y) = 1   
#  y = x**2 – 1
%matplotlib inline
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (np.exp(x) - 1)/x +1)
plt.plot(x, x**2 - 1)
plt.ylim(-1,8) 
plt.grid(True)
plt.axis('scaled')
plt.show()

from scipy.optimize import fsolve

def equations(vars):
    x, y = vars
    eq1 = x**2 - 1 - y
    eq2 = np.exp(x) + x * (1 - y) - 1
    return (eq1, eq2)

x1, y1 =  fsolve(equations, (2, 3))
x2, y2 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
```

```python
#решить систему неравенств:
#  exp(x) + x∙(1 – y) > 1   
#  y = x**2 – 1
%matplotlib inline
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, 201)

plt.figure(figsize = (5, 8))
plt.plot(x, (- 1/x)+2)
plt.plot(x, x**2 - 1)
plt.ylim(-1,6) 
plt.grid(True)

plt.show()

from scipy.optimize import fsolve

def equations(vars):
    x, y = vars
    eq1 = x**2 - 1 - y
    eq2 = (- 1/x)+2 - y
    return (eq1, eq2)

x1, y1 =  fsolve(equations, (1, 2))
x2, y2 =  fsolve(equations, (0.5, 0.5))
x3, y3 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
