```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-35, 35, 12)
Y = np.arange(-35, 35, 12)
X, Y = np.meshgrid(X, Y)
Z2 = 8 - 4*X - 6*Y
Z1 = -4*(X+20) - (Y+40) * 6
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
show()
```
