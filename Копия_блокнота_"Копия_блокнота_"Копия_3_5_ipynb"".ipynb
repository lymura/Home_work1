```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000)
y1 = np.sqrt(((x**2 - 1)/4**2)/3**2)
y2 = -(np.sqrt(((x**2 - 1)/4**2)/3**2))
plt.ylim(-1.5, 1.5)
plt.xlim(-4, 4)
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
x = np.linspace(-5, 5, 1000)
y1 = np.sqrt((1 - (x**2)/4**2)/3**2)
y2 = -(np.sqrt((1 - (x**2)/4**2)/3**2))
plt.ylim(-1,5, 1,5)
plt.xlim(-4, 4)
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
x = np.linspace(-5, 5, 1000)
y1 = np.sqrt((6 - x**2)/3)
y2 = -(np.sqrt((6 - x**2)/3))
plt.ylim(-1.5, 1.5)
plt.xlim(-4, 4)
plt.plot(x, y1)
plt.plot(x, y2)
```
