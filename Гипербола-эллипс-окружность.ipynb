```python
#уравнение гиперболы
from sympy import plot_implicit, Eq
from sympy.abc import x, y

a = -5
b = 1
plot_implicit(Eq((y - a * x) * (y - b * x), -1), (x, -10, 10), (y, -10, 10))
```

```python
#Уравнение эллипса
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 1000)
y1 = np.sqrt((1 - (x**2)/4**2)/3**2)
y2 = -np.sqrt((1 - (x**2)/4**2)/3**2)
plt.ylim(-1,5, 1,5)
plt.xlim(-4, 4)
plt.plot(x, y1)
plt.plot(x, y2)
```
