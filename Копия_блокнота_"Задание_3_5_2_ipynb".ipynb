```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
# сфера
fig = plt.figure(figsize=(8,6),facecolor='white')

ax = fig.add_subplot(2, 2, 1, projection='3d')
u = np.linspace(0, 2 * np.pi, 51)
v = np.linspace(-np.pi/2, np.pi/2, 51)
x = np.outer(np.cos(u), np.cos(v))
y = np.outer(np.sin(u), np.cos(v))
z = np.outer(np.ones(np.size(u)), np.sin(v))
ax.plot_surface(x, y, z, rstride=1, cstride=1, linewidth=0,color=(0.5, 0.5, 1))
#парабола
fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z1 =  np.sqrt(5.*(X**2 + Y**2)/1. + 2)
xcolors = X - min(X.flat)
xcolors = xcolors/max(xcolors.flat)
ax.plot_surface(X, Y, Z1, facecolors=cm.hot(xcolors))
plt.show()
```
